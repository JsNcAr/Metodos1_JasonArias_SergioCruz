```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
def espiral(a, b, theta):
    return a + (b * theta)

def r_to_cartesian(r, theta):
    return r * np.cos(theta), r * np.sin(theta)
```

## a)

```python
a = 0
b = 1
theta = np.linspace(0, 2*np.pi, 100)
r = espiral(a, b, theta)
print(r)
```

## b)

```python
theta = np.linspace(0, 6*np.pi, 100)
r = espiral(a, b, theta)

x, y = r_to_cartesian(r, theta)
```

```python
plt.plot(x, y, 'b-', linewidth=2, label='Espiral de arquimedes')
plt.legend() 
plt.show()
```
