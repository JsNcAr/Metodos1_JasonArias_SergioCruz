```python
import numpy as np 
import matplotlib.pyplot as plt
```

```python
def derivada (f,x,h):
    return (f(x+h)- f(x))/h


def sin(x): 
    return np.sin(x)
    
```

```python
h=0.1
x= np.linspace (0, 2*np.pi,num=100)
```

```python
y_c= derivada(sin,x,h)
plt.scatter (x,y_c, s=10, marker='x')
plt.plot(x,np.cos(x), color="green")
plt.show()
```
