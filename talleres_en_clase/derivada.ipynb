```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
def der(f, x, h):
    return (f(x + h) - f(x))/h


def sin(x): return np.sin(x)
```

```python
h = 0.1
x = np.linspace(0, 2*np.pi, num=100)
```

```python
y_d = der(sin, x, h)
plt.scatter(x, y_d, s=10, marker='x')
plt.plot(x, np.cos(x), color="green")
plt.show()
```

```python
def izq(f, x, h):
    return (f(x) - f(x-h))/h
```

```python
y_i = izq(sin, x, h)
plt.scatter(x, y_i, s=10, marker='x')
plt.plot(x, np.cos(x), color="red")
plt.show()
```

```python
def central(f, x, h):
    return (f(x+h) - f(x-h))/(2*h)
```

```python
y_c = central(sin, x, h)
plt.scatter(x, y_c, s=10, marker='x')
plt.plot(x, np.cos(x), color="red")
plt.show()
```

```python
print(f'El error por derecha es {np.mean(np.cos(x)-y_d)}')
print(f'El error por izquierda es {np.mean(np.cos(x)-y_i)}')
print(f'El error central es {np.mean(np.cos(x)-y_c )}')
```
