## a) $P(\theta) = 0$

Dado el segundo axioma de Kolmogórov que dice que P(W) = 1. La probabilidad del espacio de sucesos elementales W es 1

Si tenemos un evento fuera del espacio de sucesos elementales, la probabilidad de que este ocurra debe ser núlo puesto que 

\begin{align*}
    &P(W) + P(\theta) = 1 \\
    &P(W) - 1 + P(\theta) = 0 \\
    &P(\theta) = 0
\end{align*}

## b) $P(A^{c}) = 1 - P(A)$

Dado el tercer axioma de Kolmogórov que dice que la probabilidad de la unión de todos ellos es igual a la sumatoria de las probabilidades de todos los sucesos. Tenemos que para dos sucesos

\begin{align*}
    &P(W) = 1 \\
    &P(A) + P(B) = 1 
\end{align*}

Pero los sucesos se pueden escribir en términos de uno solo dado que $P(B) = P(A^{c})$
Luego

\begin{align*}
    &P(A) + P(A^{c}) = 1 \\
    &P(A^{c}) = 1 - P(A)
\end{align*}

## f) $P(A \cup B) = P(A) + P(B) - P(A \cap B) $ 



Con el diagrama de Venn tenemos que lo que queremos calcular es la probabilidad unida de los eventos A y B

Por lo cual podemos sumar la probabilidad $P(A)$ y $P(B)$ para este proposito, sin embargo, al hacer esto nos damos cuenta que estamos sumando dos veces la intersepción, inter en el diagrama, de estos dos eventos por lo cual es necesario restar una vez $P(A \cap B)$ Para eliminar ese elemento repetido y así finalmente tener que 

$P(A \cup B) = P(A) + P(B) - P(A \cap B) $
