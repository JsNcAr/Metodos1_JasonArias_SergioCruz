```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
A = 1
n_sides = 5
d = [0, np.pi/4, np.pi/2]
theta = np.linspace(- 2*np.pi, 2*np.pi, 100)
```

```python
k = 0
for i in range(2, 6):
    for j in range(1, 6):
        if i > j:
            for d_ in range(len(d)):
                k += 1
                try:
                    fig = plt.subplot(n_sides+1, n_sides+1, k)
                    fig.plot(A*np.cos(j*theta), A*np.sin(i*theta+ d[d_]))
                    fig.axes.set_aspect('equal')
                    fig.axis('off')
                except:
                    print('Error')
                
```
