```python
%matplotlib inline
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
data = pd.read_csv('./EstrellaEspectro.txt', header=None, sep='\s+')
x = data[0]
y = data[1]
plt.plot(x, y)
plt.show()
```

```python
def maximos_locales(data):
    criticos = []
    for i in range(1, max(data.index)):
        if y[i] > y[i+1] and y[i] > y[i-1]:
            criticos.append(i)
    return (criticos)
```

```python
local_max = maximos_locales(data)
```

```python
plt.plot(x, y)
plt.scatter(x[local_max], y[local_max], c='r', marker='x', s=10)
plt.show()
```
